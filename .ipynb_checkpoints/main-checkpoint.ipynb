```python
import pandas as pd
import threading
import matplotlib.pyplot as plt
import seaborn as sns

# Função para processar cada arquivo e extrair os dados relevantes
def processar_arquivo(nome_arquivo, dados_por_idade):
    with open(nome_arquivo, "r") as arquivo:
        # Extrair o ano do nome do arquivo
        ano = int(nome_arquivo.split("\\")[-1][3:7])

        for pos in arquivo:
            # Determinar a posição dos dados de interesse com base no formato do arquivo para cada período
            # 'MAGER' e 'DBWT' mudam de posição
            if 2007 <= ano <= 2013:
                idade_mae = int(pos[88:90])
                peso_bebe = int(pos[462:466])
            elif 2014 <= ano <= 2022:
                idade_mae = int(pos[74:76])
                peso_bebe = int(pos[503:507])
            else:
                raise ValueError("Ano de arquivo não suportado.")

            if 13 <= idade_mae <= 19:
                # Verificar se a idade da mãe está no intervalo desejado
                if idade_mae not in dados_por_idade:
                    dados_por_idade[idade_mae] = {
                        "peso_total": peso_bebe,
                        "quantidade": 1,
                    }
                else:
                    dados_por_idade[idade_mae]["peso_total"] += peso_bebe
                    dados_por_idade[idade_mae]["quantidade"] += 1

# Lista de arquivos a serem processados
arquivos = [
    r"datasets\Nat2007us\Nat2007us.dat",
    r"datasets\Nat2008us\Nat2008us.dat",
    r"datasets\Nat2009us\Nat2009usPub.r20131202",
    r"datasets\Nat2010us\Nat2010PublicUS.r20131202",
    r"datasets\Nat2011us\Nat2011PublicUS.r20131211",
    r"datasets\Nat2012us\Nat2012PublicUS.r20131217",
    r"datasets\Nat2013us\Nat2013PublicUS.r20141016",
    r"datasets\Nat2014us\Nat2014PublicUS.c20150514.r20151022.txt",
    r"datasets\Nat2015us\Nat2015PublicUS.c20160517.r20160907.txt",
    r"datasets\Nat2016us\Nat2016PublicUS.c20170517.r20190620.txt",
    r"datasets\Nat2017us\Nat2017PublicUS.c20180516.r20180808.txt",
    r"datasets\Nat2018us\Nat2018PublicUS.c20190509.r20190717.txt",
    r"datasets\Nat2019us\Nat2019PublicUS.c20200506.r20200915.txt",
    r"datasets\Nat2020us\Nat2020PublicUS.c20210506.r20210812.txt",
    r"datasets\Nat2021us\Nat2021US.txt",
    r"datasets\Nat2022us\Nat2022PublicUS.c20230504.r20230822.txt",
]

# Dicionário para armazenar os dados por idade da mãe
dados_por_idade = {}

print("Iniciando a extração e processamento dos dados...\n")

threads = []

# Iniciar uma thread para cada arquivo para processamento paralelo
for arquivo in arquivos:
    thread = threading.Thread(target=processar_arquivo, args=(arquivo, dados_por_idade))
    thread.start()
    threads.append(thread)

# Aguardar a conclusão de todas as threads antes de prosseguir
for thread in threads:
    thread.join()

# Criar uma lista de dicionários contendo a idade da mãe e o peso médio dos recém-nascidos
dados = []
for idade in range(13, 20):
    if idade in dados_por_idade:
        dados_idade = dados_por_idade[idade]
        peso_total = dados_idade["peso_total"]
        quantidade = dados_idade["quantidade"]
        peso_medio = peso_total / quantidade if quantidade != 0 else 0
        dados.append({"idade_mae": idade, "peso_medio": peso_medio})
        print(
            f"idade: {idade} ; peso total: {peso_total} g ; peso médio: {peso_medio:.1f} g ; n: {quantidade}"
        )

# Converter os dados em um DataFrame do pandas
df = pd.DataFrame(dados)

# Calcular a correlação entre a idade da mãe e o peso médio dos bebês
correlacao = df["idade_mae"].corr(df["peso_medio"])

print(f"\nCorrelação entre idade da mãe e peso médio dos bebês: {correlacao:.2f}\n")

sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(x='idade_mae', y='peso_medio', data=df, scatter_kws={'s': 100})
plt.title('Correlação entre Idade da Mãe e Peso Médio dos Recém-Nascidos')
plt.xlabel('Idade da Mãe')
plt.ylabel('Peso Médio dos Recém-Nascidos (g)')
plt.grid(True)
plt.show()
```
